```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Establish a connection to the database
connection = sqlite3.connect('airbnb.db')
cursor = connection.cursor()
```

```python
# Read sqlite query results into a pandas DataFrame
airbnbListings = pd.read_sql_query("SELECT * from listings", connection)
airbnbListings = airbnbListings.drop_duplicates(subset='id', keep="first")
# List the columns in the listings table
print(airbnbListings.columns)
```

## availability_30

The numerical variable availability_30 describes the number of available nights for a given listing within the next month. We should expect to see a minimum near 0, meaning the listing was fully booked for the next 30 days. The maximum should be 30, meaning the listing was not booked at all in the next 30 days. I would expect the average to be around 10, which means the listing was booked 66% of the time which gives an owner enough time to fix the house up before another group of people stay.

```python
airbnbListings.availability_30.describe()
```

As we can see, the values here closely align with what we anticipated. The minimum number of days is near 0, the max is 30 and the mean is near 10. Let's visualize the data to get an idea of the densities of each value.

```python
figure = plt.figure(figsize=(10, 6))
data = airbnbListings.availability_30
binwidth = 5
axes = figure.add_subplot(1, 1, 1)
axes.hist(data, color="darkslategray", bins=range(min(data), max(data), binwidth))
axes.set_title("Distribution of availability_30")
axes.set_xlabel("availability_30")

plt.show()
plt.close()
```

As the data above would show, a majority of listings cycle tenants quickly. Meaning there is only 5 days that are not booked within the month. This could prove to be effective in our final model. It would make sense that the best litings are the ones that are rented out the most frequent. This would lead to lower availability.

## calendar_last_scraped

The categorical variable calendar_last_scraped denotes what day that the data for a listing was scraped. Since I would believe the data was collected in batches, I would expect that there are only a few number of unique dates. Since our data covers the last 4 quarters, I expect all the dates to be within the past year.

```python
airbnbListings.calendar_last_scraped.describe()
```

It would appear that our assumptions were correct. There are only 6 possible unique dates. This makes sense because we only used a few excel sheets to read in the data. It would make sense if they were scraping data everyday within the past year. That would seem inefficient. Let's actually visualize the distribution of the data.

```python
figure = plt.figure(figsize=(10, 6))
data = airbnbListings.calendar_last_scraped

axes = figure.add_subplot(1, 1, 1)
axes.hist(data, color="darkslategray")
axes.set_title("Distribution of availability_90")
axes.set_xlabel("availability_90")

plt.show()
plt.close()
```

As we can see, our 6 unique dates are 3/19/23, 3/31/23, 12/20/22, 9/14/22, 9/15/22, 6/11/22. I find it that the most recent date does not have the highest density. The highest density is 3/19/23 which does not make too much sense to me. This might require further analysis to proceed. This variable does not seem to be a useful variable for our model.

# Pairwise EDA

## Price v. availability_30
As we have previously found, there are some listings that have no avaialability in the next 30 days, and there are some listings that have full availability in the next 30 days. I would expect that there may be a correlation between the two values. Maybe the best pricing options are all taken so that the higher availability would mean the higher prices.

```python
figure = plt.figure(figsize=(10, 6))

axes = figure.add_subplot(1, 1, 1)
axes.scatter(airbnbListings.availability_30, airbnbListings.price, marker="o", color="darkslategray")
y = [0, 250, 500, 750]
ticks = ['$0', '$250', '$500', '$750']
plt.yticks(y, ticks)
axes.set_ylabel("Price")
axes.set_xlabel("Avaialability 30")
axes.set_title("Scatter Plot of Price vs. Avaialability 30")

plt.show()
plt.close()
```

There does not seem to be any correlation between availability and price. The data seems evenly scattered with no strong relationships. I was expecting to see some clustering or monotonic relationshiops but this is not the case.
